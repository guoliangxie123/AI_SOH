```python
def charging_energy(start, end, data):
    time_delta = (pd.to_datetime(end) - pd.to_datetime(start)).seconds
    charging_delta = data[(data['数据时间'] >= start) & (data['数据时间'] <= end)]
    sum_energy = (20/ 3600) * abs(sum(charging_delta['总电流']))
    return sum_energy
```

```python
k = 0
for i in tqdm(range(len(data))):

    if data.iloc[i]['充电状态'] == 1 and data.iloc[i - 1]['充电状态'] != 1:
        k += 1
        data.at[i, '充电周期'] = k
    elif data.iloc[i]['充电状态'] == 1 and data.iloc[i - 1]['充电状态'] == 1:
        data.at[i, '充电周期'] = k
    else:
        continue
```

```python
# 充电周期
k=0
for i, row in tqdm(data.iterrows()):
    if row['充电状态'] == 1:
        data.loc[i,'charge_no'] = k
    elif i == len(data)-1:
        break
    elif row['充电状态'] != 1 and data.loc[i+1,'充电状态'] ==1:
        k+=1
    else:
        continue
```

```python
k = data[data['充电周期'] >= 1]['充电周期'].nunique()
data_train = pd.DataFrame()
for i in range(1, k + 1):
    soc_max = data[data['充电周期'] == i]['SOC'].max()
    soc_min = data[data['充电周期'] == i]['SOC'].min()
    print(i,soc_max,soc_min)
    if soc_max >= 90 and soc_min <= 40:
        data_train = pd.concat([data_train, data[(data['充电周期'] == i) & (data['SOC'] >= 40) & (data['SOC'] <= 90)]])
```

```python
k1 = data_train[data_train['充电周期'] != -1]['充电周期'].unique().tolist()
values = []
for i in k1:
    start = data_train[data_train['充电周期'] == i]['数据时间'].min()
    end = data_train[data_train['充电周期'] == i]['数据时间'].max()
    time_delta = (pd.to_datetime(end) - pd.to_datetime(start)).seconds
    mile = data_train[data_train['充电周期'] == i]['累计里程'].min()
    energy = charging_energy(start,end,data_train[data_train['充电周期'] == i])
    values.append([start,end,time_delta,mile,energy])
values
```
